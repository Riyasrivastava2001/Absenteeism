```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
fig, ax=plt.subplots(figsize=(10,7))


ax.bar(raw_data["ID"],
      raw_data["Reason for Absence"],
      color="k")
ax.set_ylabel("Reason For Absence")

ax1=ax.twinx()

ax1.plot(raw_data["ID"],
      raw_data["Absenteeism Time in Hours"],
         color="#b60000",
        marker="D")
ax1.set_ylabel("Absenteeism Time in Hours",
              color="#b60000")
plt.title("Survey of People Absent")
```

```python
plt.figure(figsize=(9,6))
plt.bar(x=df2["Absenteeism Time in Hours"],
       height=df2["Date"],
       color=["brown"])
plt.xticks(fontsize=9,fontweight="bold")
plt.yticks(fontsize=9,fontweight="bold")
plt.title("Date of Absenteeism",fontsize=12,fontweight="bold")
plt.ylabel("Date",fontsize=11,fontweight="bold")
plt.xlabel("Absenteeism Time in Hours",fontsize=11,fontweight="bold")
plt.show()
```

```python
plt.figure(figsize=(20,8))
labels=("Reason1","Reason2","Reason3","Reason4","Absenteeism Time in Hours")
plt.plot(df2["Absenteeism Time in Hours"],df2["Reason1"])
plt.plot(df2["Absenteeism Time in Hours"],df2["Reason2"])
plt.plot(df2["Absenteeism Time in Hours"],df2["Reason3"])
plt.plot(df2["Absenteeism Time in Hours"],df2["Reason4"])
plt.title("Reasons for Absenteeism",fontsize=18,fontweight="bold")
plt.xlabel("Absenteeism Time in Hours",fontsize=14,fontweight="bold")
plt.ylabel("Reasons",fontsize=14,fontweight="bold")
plt.legend(labels=labels,fontsize=18)
plt.show()
```
